```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data_dir = "../Data/Raw/"
data_2016 = pd.read_csv(data_dir+"roc_2016.csv")
data_2016.head()
```

```python
data_2017 = pd.read_csv(data_dir+"roc_2017.csv")
data_2017.head()
```

```python
np.savetxt(r'../Data/Untokenized/train.src.en', X_train.values, fmt="%s")
np.savetxt(r'../Data/Untokenized/train.tgt.en', y_train.values, fmt="%s")
```

### Validation Data

```python
validation_data = pd.read_csv(data_dir+'validation.csv')
print("Validation Data: %i" %len(validation_data))
validation_data.head()
```

```python
X_valid = pd.DataFrame(columns=["Input"])
y_valid = pd.DataFrame(columns=["Correct Ending"])
X_valid["Input"] = validation_data["InputSentence1"]+" "+validation_data["InputSentence3"]+" "+validation_data["InputSentence4"]
X_valid.head()
```

```python
X_valid["Input"] = "<S> "+X_valid["Input"]+" </S>"
X_valid.head()
```

```python
correct_endings = []
for row in range(len(validation_data)):
    if (int(validation_data.iloc[row]["AnswerRightEnding"]) == 1):
        correct_endings.append(validation_data.iloc[row]["RandomFifthSentenceQuiz1"])
    elif (int(validation_data.iloc[row]["AnswerRightEnding"]) == 2):
        correct_endings.append(validation_data.iloc[row]["RandomFifthSentenceQuiz2"])
y_valid["Correct Ending"] = correct_endings
y_valid["Correct Ending"] = "<S> "+y_valid["Correct Ending"]+" </S>"
print(len(y_valid))
y_valid.head()
```

```python
np.savetxt(r'../Data/Untokenized/valid.src.en', X_valid.values, fmt="%s")
np.savetxt(r'../Data/Untokenized/valid.tgt.en', y_valid.values, fmt="%s")
```
